# 다변량 그래프 그리기

:3개 이상의 변수를 사용허ㅏㄴ 그래프프

```python
#1.앞서 실습한 산점도 그래프(scatter„plot)에 성 별을 새 번수로 추가
#여성 (Female)인 경우 0을 반환하고 남성 (Male)인 경우 1을 반환하는 함수

def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1

#2.recode_sex 메서드가 반환한 값(0,1)을 데이터프레임에 추가
#sex 열에 recode_sex 함수'^ L로드캐스팅하기 위하여 apply 메서드를 사용

tips['sex_color'] = tips['sex'].apply(recode_sex)

#3.테이볼당 인원 수(size)도 산점도 그래프에 추가
scatter_plot = plt.figure( ) 
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'], y=tips['tip'], s=tips['size'] * 10, c=tips['sex_color'],alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by size')
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
```

#04-3 seaborn 라이브러리 자유자재로 사용

```python
#1.
import seaborn as sns
tips = sns.load_dataset("tips")


#2.seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용.
#subplots 메서드로 기본 틀을 만듈고 distplot 메서드에 total_bill 열 데이터를 전달

ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
#3.밀집도 그래프를 제외하고 싶다면 kde 인잣값을 False로 설정
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], kde=False) 
ax.set_title('Total Bill Histogram') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Frequency')
```

```python
#4.밀집도 그래프만 나타내려면 hist 인자를 False

ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], hist=False) 
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
#5. 히스토그램과 밀집도 그리는 distplot 매서드에 rug 인자 추가해 rug 그래프까지

ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#6.히스토그램과 비슷하게 생긴 count 그래프. count 그래프는 이산값을 나타낸 그래프
ax = plt.subplots( )
ax= sns.countplot('day', data=tips)
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency')
```

```python
#1. seaborn 라이브러리로 산점도 그래프 그리기
#   : seaborn 라이브러리의 regplot 메서드,산점도 그래프와 회귀선을 함께(회귀선을 제거하려면 fit_reg 인자를를 False) 

ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')


ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#2. 산점도 그래프와 히스토그램을 한 번에 그려주는 jointplot 메서드

joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)


#3.산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점,
#:  구분하기 쉽게 그리고 싶다면 육각 그래프/ jointplot 메서드를를 그대로 사용. 대신 kind 인잣값을 hex
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#4. 이차원 밀접도 그리기 : kdeplot
#     shade 인잣값을 True로 지정하면 그래프에 음영 효과
ax = plt.subplots( )
ax= sns.kdeplot(data=tips['total_bill'],,data2=tips['tip'],shade=True)
ax.set_title('Kernel Density Plot if Total Bill and Tip')
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
#5. 바 그래프 그리기 : barpot

ax = plt.subplots( )
ax = sns.barplot(x='time', y='total_bill', data=tips) 
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

```python
#6. 박스 그래프 그리기

#최솟값, 1분위수, 중간값, 3분위수, 최댓값,이상치 등 다양한 통계량을 한 번에

ax = plt.subplots( )
ax = sns.boxplot(x='time', y='total_bill', data=tips) 
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
#7. 바이올린 그래프 :박스형태를 바이올린 형태로 바꿈  : sns.violinplot
ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', data=tips) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
#8. 관계그래프 그리기   sns.pairplot
#   : 지금까지 설명한 그래프를 종합한 그래프

#단점 : 대각선기준으로 중복된 그래프 그려짐.

fig= sns.pairplot(tips)
```

```python
#9. map_upper/ map_lower/ map_diag

pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) 
plt.show()
```

#FacetGrid 클래스로 그룹별 그래프 만들기

```python
#1. 먼저 FaceGrid에 데이터프레임과 그룹을 구분할 열 이름을 전달해 객체 생성.
#   객체의 map 메서드에 그래프의 종류와 x축으로 지정할 열 이름을 순서대로 전달하면 히스토그램, 밀집도, 양탄자 그래프 그릴 수 있음

facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill', rug=True)
```

#04-5 seaborn라이브러리로 그래프 스타일 설정



```python
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
#2. whitegrid로 스타일 지정   :sns.set_style
sns.set_style('whitegrid') 
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)

```

```python
#3. for문 활용한 스타일 적용

fig = plt.figure( )

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
